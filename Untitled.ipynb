```python
import numpy as np
import os
from cmaes import CMA
from visualizer.csa import CMA_CSA
from bisect import bisect_left

from matplotlib.pyplot import plot 
```

```python
def get_percentile(score, score_list):
    score_list.sort(reverse = True) # asc
    index = easy_search(score, score_list)
    print(index)
    return (index / len(score_list)) * 100
    
        

def easy_search(score, score_list):
    
    for i in range(0, len(score_list)):
        if score > score_list[i]:
            return i
    return len(score_list)
            
            
```

```python

def quadratic(x1, x2):
    return (x1 - 3) ** 2 + (10 * (x2 + 2)) ** 2


def main():
    optimizer = CMA_CSA(mean=np.array([-1000.0, 280.0]), sigma=1.3, population_size = 10)
    print(" g    f(x1,x2)     x1      x2  ")
    print("===  ==========  ======  ======")
    
    expert_qualities = []

    while True:
        solutions = []
        values = []
        for _ in range(optimizer.population_size):
            x = optimizer.ask()
            value = round(quadratic(x[0], x[1]), 5)
            values.append(value)
            solutions.append((x, value))
            print(
                f"{optimizer.generation:3d}  {value:10.5f}"
                f"  {x[0]:6.2f}  {x[1]:6.2f}"
            )
        optimizer.tell(solutions)
        
        expert_score = quadratic(optimizer._mean[0], optimizer._mean[1])
        expert_percentile = get_percentile(expert_score, values)
        expert_qualities.append(expert_percentile)
        print(f'Expert location: {optimizer._mean}')
        print(f'Expert quality: {expert_score}')
        print(f'Expert better than {expert_percentile}% of population')
        if optimizer.should_stop():
            break
            
    return expert_qualities
        
    


qualities = main()

plot([i for i in range(1, len(qualities) + 1)], qualities)
```
